# Calculator

```python
import tkinter as tk

# Function to update the display
def update_display(text):
    display_var.set(text)

# Function to handle button clicks
def button_click(button_text):
    current_text = display_var.get()
    new_text = current_text + button_text
    update_display(new_text)

# Function to handle equal button click
def equal_button_click():
    try:
        result = eval(display_var.get())
        update_display(result)
    except ZeroDivisionError:
        update_display("Error")

# Function to handle clear button click
def clear_button_click():
    update_display("")

# Create the main window
root = tk.Tk()
root.title("Calculator")

# Create a StringVar to store the display text
display_var = tk.StringVar()
display_var.set("")

# Create the display label
display_label = tk.Label(root, textvariable=display_var, font=("Helvetica", 24), anchor="e", padx=10, pady=10)
display_label.grid(row=0, column=0, columnspan=4)

# Create the buttons
buttons = [
    ["7", 1, 0],
    ["8", 1, 1],
    ["9", 1, 2],
    ["/", 1, 3],
    ["4", 2, 0],
    ["5", 2, 1],
    ["6", 2, 2],
    ["*", 2, 3],
    ["1", 3, 0],
    ["2", 3, 1],
    ["3", 3, 2],
    ["-", 3, 3],
    ["0", 4, 0],
    [".", 4, 1],
    ["=", 4, 2],
    ["+", 4, 3],
    ["C", 5, 0]
]

for button in buttons:
    text = button[0]
    row = button[1]
    col = button[2]
    if text == "=":
        button = tk.Button(root, text=text, font=("Helvetica", 24), command=equal_button_click)
    elif text == "C":
        button = tk.Button(root, text=text, font=("Helvetica", 24), command=clear_button_click)
    else:
        button = tk.Button(root, text=text, font=("Helvetica", 24), command=lambda t=text: button_click(t))
    button.grid(row=row, column=col, sticky="nsew")

# Configure grid weights for resizing
for i in range(6):
    root.grid_rowconfigure(i, weight=1)
for i in range(4):
    root.grid_columnconfigure(i, weight=1)

# Start the main event loop
root.mainloop()
```
